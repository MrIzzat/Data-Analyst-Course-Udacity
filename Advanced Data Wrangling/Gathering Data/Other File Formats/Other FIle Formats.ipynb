# What other file formats are used?

There are many other file formats that are compatible with pandas such as:

* Excel
* Pickle
* HDF5
* SAS
* SATA

Which file format should be used? It depends on the type of data, like flat files being food for tabular data.

For Image interaction, the Pillow library is pretty good for interacting with images.

Non-relational databases can store many different types of data such as images and videos and text files. It can store structured and unstructured data.

MongoDB is a NoSQL database.

In MongoDB, for relational data:

* Allows to embed documents within other documents.
* Stores relationships between data by creating links between the data.

MongoDB allows for storing and retrieving unstrucuted data, like a JSON file. It uses an alternative version of JSON called BSON which is basically Binary JSON. An advantage of BSON is it can be traversed much more quickly than JSON as well as supporting more advanced types of data like integers vs floats which JSON cannot do.

# Import Libraries

```python
from PIL import Image
```

# Load Image

```python
image = Image.open('goldfish.jpg')
display(image) #image.show() opens the picture on my computer
```

```python
image.size
```

# Flipping the image

```python
flipped_image = image.transpose(Image.FLIP_TOP_BOTTOM)
display(flipped_image)
```

# Saving the flipped image

```python
flipped_image = flipped_image.save('flipped_goldfish.jpg')
```
