# Import Libraries

```python
import pandas as pd
```

# Read Data

```python
df = pd.read_csv('data_08_v2.csv')
df.head(1)
```

# Query String Data

## Getting all cars that are two-wheel drive

### Using df[]

```python
df_2wd = df[df["drive"]=="2WD"]
df_2wd.head()
```

### Using .query()

```python
df_2wd = df.query('drive == "2WD"')
df_2wd.head()
```

## Select records of people in California

### Using df[]

```python
df_cert_ca = df[df["cert_region"]=='CA']
df_cert_ca.head()
```

### Using .query()

```python
df_cert_ca = df.query(' cert_region == "CA"')
df_cert_ca.head()
```

# Query Numeric Data

## Finding the displacement lower than 3

### Using df[]

```python
df_low_displ = df[df.displ < 3]
df_low_displ.head()
```

### Using .query()

```python
df_low_displ = df.query('displ < 3')
df_low_displ.head()
```
