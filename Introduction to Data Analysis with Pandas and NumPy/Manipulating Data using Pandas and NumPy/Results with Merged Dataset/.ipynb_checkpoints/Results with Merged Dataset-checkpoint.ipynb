## Getting the maximum difference between the combined milaget from 2008 and 2018

Basically what is done is that every value from the 2018 column is subtracted from the 2008 column and placed into a series (a single column). The id of the maximum value is then returned using `idxmax()`

```python
df.iloc[( df['cmb_mpg'] - df['cmb_mpg_2008']).idxmax()]
```
