# Goals

* Visualization the relationship between temperature and electrical output
* Visualize the disttribution of humdity
* Visualize the feature with the most outliers

# import libraries

```python
import pandas as pd
```

# Load Dataset

```python
df = pd.read_csv('powerplant_data_edited.csv')
```

```python
df.head(1)
```

# Plot the relationship between temperature and electrical output

```python
df.plot(kind='scatter', x='temperature', y='energy_output');
```

# Plot the ditribtution of humidity using a histogram

```python
df['humidity'].value_counts().plot(kind='hist');
```

# Plotting Box Plots

## `Temperature`

```python
df['temperature'].value_counts().plot(kind='box');
```

## `exhaust_vacuum`

```python
df['exhaust_vacuum'].value_counts().plot(kind='box');
```

## `pressure`

```python
df['pressure'].value_counts().plot(kind='box');
```

## `humidity`

```python
df['humidity'].value_counts().plot(kind='box');
```

## `energy_output`

```python
df['energy_output'].value_counts().plot(kind='box');
```
