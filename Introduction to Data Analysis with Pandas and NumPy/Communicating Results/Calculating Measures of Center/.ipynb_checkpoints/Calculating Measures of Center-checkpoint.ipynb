# What are measures of center?

When talking about measures of center, the subjects that are meant are:

* Mean: Is basically the average, Example: [1, 2, 3, 4, 5] has an average of 3 (15 divided by 5).

* Median: Is the center value of a set after sorting. If the number of valuesi s even then the midpoint is taken between the two center values.Example:[2, 1, 4, 5, 3] sorts to [1, 2, 3, 4, 5], 3 is the median. While [2, 1, 4, 5, 3, 6] sorts to [1, 2, 3, 4, 5, 6], 3.5 is the median

* Mode: Is the value with the highest frequency in a set: Example: 
[1, 2, 2, 3, 3, 3, 4, 4, 4, 4], 4 is the mode.

Measures of center help provide a summary of a dataset that is focused on its center. Measurements for center can provide a quick and easy way to know the basic characteristics of a dataset. They are also very helpful for calculatng other analytics seach as spread and outliers.

# Import libraries

```python
import pandas as pd
```

# Load Dataset

```python
df = pd.read_csv('census_income_data.csv')
```

# Using `mean()` on a dataframe

Using mean on `capital-gain` and `capital-loss` to answer: 

    "what was the average capital gained and lost in the dataset?"

```python
df[["capital-gain","capital-loss"]].mean()
```

# Using `mean()` and `median()` on a group

## On the `workclass` group

This can be applied for a question like:

    "What is the average and median for all other values for each workclass?"

First, it's good to know what are the workclass types

```python
df["workclass"].value_counts()
```

### Finding the mean

```python
df.groupby('workclass').mean(numeric_only=True)
```

### Finding the median

```python
df.groupby('workclass').median(numeric_only=True)
```

## On the occupation group

Answering the question:

    "What is the average and median for all other values for each occupation?"

```python
df["occupation"].value_counts()
```

### Getting the mean

```python
df.groupby('occupation').mean(numeric_only=True)
```

### Getting the median

```python
df.groupby('occupation').median(numeric_only=True)
```
