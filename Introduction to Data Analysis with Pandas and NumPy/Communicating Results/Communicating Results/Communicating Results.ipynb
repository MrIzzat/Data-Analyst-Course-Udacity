# Import libraries

```python
import pandas as pd
%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('census_income_data.csv')
```

```python
df.head(1)
```

# Creating two new Dataframes depending on the income

```python
df_a = df.query('income == " >50K"')
df_b = df.query('income == " <=50K" ')
```

# Getting the value counts of `education`

To make sure the indexes are lined up on the x-axis between the two plots, `[index]` can be added to `value_counts()` before plotting.

```python
index = df_a["education"].value_counts().index
df_a['education'].value_counts()[index].plot(kind='bar');
```

```python
df_b['education'].value_counts()[index].plot(kind='bar');
```

# Ploting `workclass` with pie charts for each class

```python
index = df_a['workclass'].value_counts().index
df_a['workclass'].value_counts()[index].plot(kind="pie");
```

```python
df_b['workclass'].value_counts()[index].plot(kind="pie");
```

# Creating a histogram for `age`

```python
print(df_a['age'].describe())
df_a['age'].hist();
```

```python
print(df_b['age'].describe())
df_b['age'].hist();
```
