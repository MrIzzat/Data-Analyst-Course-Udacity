# What is `groupby()`? 

`groupby()` allows for grouping data by one or more columns, it helps with performing specific opertaions on one or more DataFrames.

# Import Libraries

```python
import pandas as pd
```

# Load Dataset

```python
df = pd.read_csv('census_income_data.csv')
```

# View the available columns

```python
df.columns
```

# Calculate hte mean for all numeric olumns

```python
df.mean(numeric_only=True)
```

# Calculate the mean using Groupby `workclass`

```python
df.groupby('workclass').mean(numeric_only=True)
```

As it can be seen below, `groupby()` will group the data in a DataFrame based on each unique value the specified column.

```python
df['workclass'].value_counts()
```

# Calculating the mean using `groupby()` with two columns

```python
df.groupby(["workclass","race"]).mean(numeric_only=True)
```

# Using `as_index=False` with multiple columns to make sure they remain non-index values.

```python
df.groupby(["workclass","race"],as_index=False).mean(numeric_only=True)
```

```python
df.groupby(["workclass","race"],as_index=False)["capital-gain"].mean(numeric_only=True)
```
