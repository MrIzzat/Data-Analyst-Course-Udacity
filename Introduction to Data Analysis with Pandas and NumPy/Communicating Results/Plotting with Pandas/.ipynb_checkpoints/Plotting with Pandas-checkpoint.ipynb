# What are ways to explore data with visuals?

After data has been cleaned, plots are a very useful tool for finding patterns. Plots can be: histograms, scatter plots etc. Plots can help determine which variables should be used in analysis. Each type of plot can be used to visualize patterns differently based on the data. Plots in pandas is just a wrapper for the matplotlib library.

## Histograms

Histograms look like this:

![](hist_plot.jpeg)\

* Are used to dislpay the distribution of numerical data.

* Divide data into bins and show the frequency of observations for each bin. Each one of the long blue rectangles is a bin.

## Bar Charts

Bar Charts look like this:

![](bar_plot.jpeg)

* Are used to compare differente categories.

* Show values as bars of different lengths.

## Pie Charts

Pie charts look like this:

![](pie_chart.jpeg)

* Pie Charts are good for showing proportions of different categories where they display each category like slices of a pie.

## Scatter Plots

Scatter Plots look like this:

![](scatter_plot.jpeg)

* Display the relationship between two numerical variables
* Plots points on a two-dimensional graph
* When there is a positive correlation between the variables, the slope increases from left to right, in resemblence to the plot of y=x function.
* When there is a negative correlation between the variables, the slope decresesfrom left to right, as in the image above.

## Box Plot

Box Plots look like this:

![](box_chart.jpeg)\

* Displays the distrubtion of numerical data
* Shows five key values:
    * Minimum
    * First Quartile
    * Median
    * Third Quartile
    * Maximum

# Import Libraries

```python
import pandas as pd
%config InlineBackend.figure_format='retina'
```

# Read Dataset

```python
df = pd.read_csv('census_income_data.csv')
df.head(1)
```

# Histogram of all columns

```python
df.hist(figsize=(10,10));
```

# Histogram of the `age` column

```python
df['age'].hist(figsize=(5,5));
```

# Histogram of `age` column using `plot()`

```python
df['age'].plot(kind='hist',figsize=(5,5));
```

# Bar chart for `education`

The bar chart needs the values to be numerical before being plotted, that's why `value_counts()` is done first before plotting. With columns that inheritly have numerical values, it can be very slow without `value_counts()`.

```python
df['education'].value_counts().plot(kind='bar');
```

# Pie Chart for `workclass`

Pie charts need numerical values to work so `value_counts()` should be used before plotting with pie charts as well.

```python
df['workclass'].value_counts().plot(kind='pie',figsize=(10,10));
```

# Plotting data from `cancer_data_edited.csv`

## Read Dataset

```python
df = pd.read_csv('cancer_data_edited.csv')
df.info()
```

## Scatter plots for all columns

```python
pd.plotting.scatter_matrix(df,figsize=(15,15));
```

## Scatter Plot of two columns: `compactness` and `concavity`

```python
df.plot(kind='scatter', y='concavity',x='compactness');
```

## Box Plot for `concave_points`

```python
df['concave_points'].plot(kind='box',figsize=(5,5));
```
