# Plots

## Graph 1

Scatter Plot of Sepal Length vs. Sepal Width Create a scatter plot to visualize the relationship between `SepalLengthCm` and `SepalWidthCm`, colored by the `Species`.

```python
x = df['SepalLengthCm']

y= df['SepalWidthCm']
```

```python
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", data=df, hue='Species');
```

## Graph 2

Facet Grid of Sepal Length vs. Petal Length. Create a Facet Grid to show the relationship between `SepalLengthCm` and `PetalLengthCm` for each species (Species).

```python
grid = sns.FacetGrid(df, col='Species')

grid.map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()

plt.show()
```

## Graph 3

Create a plot matrix (pair plot) to visualize pairwise relationships between `SepalLengthCm`, `SepalWidthCm`, `PetalLengthCm`, and `PetalWidthCm`, colored by `Species`.
