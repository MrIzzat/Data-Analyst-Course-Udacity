# Import Libraries

```python
import seaborn as sns
import pandas as pd

%config InlineBackend.figure_format='retina'
```

## Load dataset

```python
diamonds = sns.load_dataset('diamonds') #Needs internet
```

# Intial Exploration

```python
diamonds.head()
```

```python
diamonds.isna().sum()
```

a few missing values

```python
diamonds.dropna(inplace=True)
```

```python
diamonds.isna().sum()
```

# Plots

## Scatter Plot

Visualizing the relationship between the carat and price of a diamond

```python
sns.scatterplot(data=diamonds, x='carat',y='price').set_title("Effect of carats on diamond prices");
```

## Box Plot

Seeing the distribution of prices across the different types of cuts

```python
sns.boxplot(data=diamonds, x='cut',y='price').set_title("Distribution of prices by cut type");
```

# Summary

THe relationship between the `carat` of a diamond and its `price` is that the more carats it has the more it will cost, as it can be seen when comparing diamond of 0-1 carts and 1-2 carats on the scatter plot.

The price of diamonds depending on its cut goes as follows (from expensive to cheapest): Premium, Very Good Cut, Good, Fair, Ideal
