# Import Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('./titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
values = df["Pclass"].value_counts().sort_index()
values
```

```python
value_labels = []
for i in values.index:
    value_labels.append("Class "+str(i))

value_labels.sort()
value_labels
```

```python
plt.figure(figsize=(12,8))



plt.bar(value_labels,values,color=['#FF0000','#C0C0C0','#000000'])
plt.title('Count of Passengers per Class',fontsize=20)
plt.xlabel('Class',fontsize=15,labelpad=10)
plt.ylabel('Count',fontsize=15,labelpad=10)


temp = list(values)

temp.extend([0,100,200,300,400,500])

plt.gca().set_yticks(values, minor=False)

plt.show()
```

It can be deduced that most people were in the third class, almost double either class. There weren't too many people from the first or second class as compared to the third.
