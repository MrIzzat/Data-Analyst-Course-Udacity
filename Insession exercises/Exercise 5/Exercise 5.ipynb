# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Load Dataset

```python
df = pd.read_csv('./movies.csv')
```

# Creating a new clean dataframe

```python
cleaned_movies_df = df.copy()
cleaned_movie_crew_df = cleaned_movies_df[['title', 'year', 'crew']]
```

```python
cleaned_movies_df = cleaned_movies_df.drop('crew',axis=1)
```

```python
cleaned_movies_df
```

## Renaming the columns in the clean dataframe

```python
cleaned_movie_crew_df.rename(columns={'title':'movie_title','year':'movie_year'},inplace = True)
```

```python
cleaned_movie_crew_df['crew'].info()
```

# Convert every crew row into a list instead of a string (may take some time to run)

Can be done using the `eval()` function

```python
# from ast import literal_eval #  It's actually slower than eval()
cleaned_movie_crew_df['crew'] = cleaned_movie_crew_df['crew'].apply(lambda x: eval(x))
```

## Verifying if the rows have been converted to lists

```python
type(cleaned_movie_crew_df['crew'][0])
```

```python
cleaned_movie_crew_df
```

# Making each crew have their own row

Using the `json_normalize()` function, each `crew` row can be converted converted into a list of json objects

```python
cleaned_movie_crew_df['crew'] = cleaned_movie_crew_df['crew'].apply(lambda x: pd.json_normalize(x))
```

## Validating the results of `json_normalize`

```python
cleaned_movie_crew_df
```

```python
cleaned_movie_crew_df['crew'][0]
```

# Creating a new dataframe containing only the crew

## Copying the `crew` column into a dataframe

```python
crew_df = pd.DataFrame(cleaned_movie_crew_df['crew'])

crew_df
```

## Setting the index of each crew from the new dataframe column to the `crew` column as a foreign key to create a 1-1 relationship between them

```python
cleaned_movie_crew_df['crew'] = crew_df.index
cleaned_movie_crew_df['crew']
```

## Renaming the `crew` column in the original dataframe to `crew_id`

```python
cleaned_movie_crew_df.rename(columns={'crew':'crew_id'},inplace=True)
```

```python
cleaned_movie_crew_df
```
