````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")

%config InlineBackend.figure_format='retina'
````

````python
df = pd.read_csv('cleaned.csv')
````

## Distribution of Distance

The visualization shows the distribution of distance travelled by bikes during their renting period. The graph shows that most bikes travelled between 0.25 km and 2.75 km.

````python
bins = np.arange(np.around(df['distance'].min()),10, 0.25)
ax = sns.histplot(df, x='distance',bins=bins,kde=True, kde_kws=dict(clip=(bins.min(), bins.max())))
ax.set_title('Distribution of Distance')
ax.set_xlabel('Distance (KM)')
;
````

## Distribution of Duration

> You should have at least three visualizations in your presentation, but feel free to add more if you'd like!

````python
bins = np.arange(np.around(df['duration_sec'].min()),2000+50, 50)
ax = sns.histplot(df, x='duration_sec',bins=bins,kde=True, kde_kws=dict(clip=(bins.min(), bins.max())))
ax.set_title('Distribution of Duration')
ax.set_xlabel('Duration (Seconds)')
;
````

## Box Plots of Age, Distance and Duration by Gender and Renter Type

````python
fig, axes = plt.subplots(2,3);

fig.tight_layout(rect=[0, 0.03, 1, 0.94])

fig.suptitle("Box Plots of Age, Distance and Duration by Gender and Renter Type",fontsize=24);
fig.set_size_inches(15, 10);


sns.boxplot(df,x='member_gender',y='age',ax=axes[0][0]);

sns.boxplot(df,x='member_gender',y='distance',ax=axes[0][1]); 

sns.boxplot(df,x='member_gender',y='duration_sec',ax=axes[0][2]); 

sns.boxplot(df,x='user_type',y='age',ax=axes[1][0]); 

sns.boxplot(df,x='user_type',y='distance',ax=axes[1][1]); 

sns.boxplot(df,x='user_type',y='duration_sec',ax=axes[1][2]); 


axes[0][0].set_title("Age of Renters by Gender",fontsize=16);
axes[0][0].set_xlabel("Gender");
axes[0][0].set_ylabel("Age");

axes[0][1].set_title("Distance Travelled by Gender",fontsize=16);
axes[0][1].set_xlabel("Gender");
axes[0][1].set_ylabel("Distance (KM)");

axes[0][2].set_title("Rent Duration by Gender",fontsize=16);
axes[0][2].set_xlabel("Gender");
axes[0][2].set_ylabel("Duration (Seconds)");

axes[1][0].set_title("Age of Renters by Renter Type",fontsize=16);
axes[1][0].set_xlabel("Renter Type");
axes[1][0].set_ylabel("Age");

axes[1][1].set_title("Distance Travelled by Renter Type",fontsize=16);
axes[1][1].set_xlabel("Renter Type");
axes[1][1].set_ylabel("Distance (KM)");

axes[1][2].set_title("Rent Duration by Renter Type",fontsize=16);
axes[1][2].set_xlabel("Renter Type");
axes[1][2].set_ylabel("Duration (Seconds");
;
````
