**Violen plots** can be used to plot the relationship between quantative and qualitative variables. It can show summaru statistics in addtion to the data's probability distribution using KDE. It's recommended to be used in exploration rather than explanation.

Violen plots are like scatter plots when one data is qualitative and the other is quantitative.

Despite them being designed for quantative and qualitattive variables, they can be used to compare two quantitative variables.

# Import Libraries

```python
import pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'
```

# Load Datasset

```python
df = pd.read_csv('fuel-econ.csv')
```

# Example 1: Violen Plot for Fuel Efficency and Vehicle class

THe fuel effeciency `comb` is quantitative and the vehicle class `VClass` is qualitative .

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)#Make these sedan classes into a categorical data type

df['VClass'] = df['VClass'].astype(vclasses) # Convert the vehicle classes from string to categorical data type

plt.figure(figsize=(12,4))
sns.violinplot(data=df, x='VClass', y='comb');
```

# Example 2: Violen Plot without datapoitns in the interior

The inner black shape with the white line on it is a miniature box plot. it can be removed by adding the parameter `inner=None`

```python
#plt.figure(figsize=(12,4))
sns.violinplot(data=df, x='VClass', y='comb', inner=None)
plt.xticks(rotation=15)
;
```

# Example 3: Violen plot with quartile information

By adding `inner=quartile`, the violen plot will show the statistical summary about the data showing the three quartiles, the median being the one with thick dashes.

```python
sns.violinplot(data=df, x='VClass', y='comb', inner='quartile')
plt.xticks(rotation=15)
;
```

# Other `inner` options

`box` is the default

```python
sns.violinplot(data=df, x='VClass', y='comb', inner='box')
plt.xticks(rotation=15)
;
```

`point` show each point as a dot

```python
sns.violinplot(data=df, x='VClass', y='comb', inner='point')
plt.xticks(rotation=15)
;
```

`sitck` shows each point as a line

```python
sns.violinplot(data=df, x='VClass', y='comb', inner='stick')
plt.xticks(rotation=15)
;
```

# Comparison: Box and Violen Plots

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,5))

sns.boxplot(data=df, x='VClass', y='comb', color='tab:red', ax=ax1)
ax1.tick_params(axis='x',rotation=15)

sns.violinplot(data=df, x='VClass', y='comb', color='tab:red', ax=ax2)
plt.xticks(rotation=15)
;
```

A better way to compare them is make sure they both have the same y limit.

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,5))

sns.boxplot(data=df, x='VClass', y='comb', color='tab:red', ax=ax1)
ax1.tick_params(axis='x',rotation=15)

sns.violinplot(data=df, x='VClass', y='comb', color='tab:red', ax=ax2)
plt.xticks(rotation=15)
plt.ylim(ax1.get_ylim())
;
```

It can be seen that the box plot inside the violen plot matches up exactly with the box plots created on the left.
