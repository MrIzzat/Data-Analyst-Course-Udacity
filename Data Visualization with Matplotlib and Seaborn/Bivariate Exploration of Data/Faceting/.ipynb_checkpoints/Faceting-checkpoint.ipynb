**Faceting** is agreat visualization technique useful for handling plots of two or mote variables. In faceting, data is divided into disjoint subsets, usually by different levels of categorical data. For each subset, the same plot table is rendered on other variables. Faceting is used to compare distributions or relationships across levels of additional variables, especially when there are three or more variables. Even though Faceting is most useful in multivariate visualizations, it can be very usevial in bivariate visualizations.

Instead of plotting a relationship between one numerica varaible and one categorical variable using a box plot or violen plot, faceting can be used to look at a historgram of numerica variables for the subsets of the datqa  divided by categorical variable levels.


In short, faceting can be used to plot two categorical variables in the context of numeric values. 

It can be viewed as a modified form the the clustered bar charts where the clustered bar charts plot the different levels in the same plot using different encoding shapes or colors for each bar, while faceting splits the single chart into multiple charts.

![](./Facet%20Grid%20vs%20Clustered%20Bar%20Charts.png)

Faceting is most useful when comparing categorical data with a lot of levels (unique categorical values).

Axis scales and limits are important to consider when plotting with facet grids.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format ='retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

## Converting the `VClasss` column into an ordered categorical type

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)
df['VClass'] = df['VClass'].astype(vclasses);
```

# Example 1

Plotting a facet grid using seaborn's `FacetGrid()`

```python
g = sns.FacetGrid(data=df, col='VClass')
g.map(plt.hist, 'comb');
```

# Example 2

Cleaning up the bin edges

```python
bin_edges = np.arange(12, 58+2, 2)

g = sns.FacetGrid(data=df, col='VClass')#sharey=False can be used to limit the y axis to the max value of the plot
                                        #col_wrap=3 can be used to show at most 3 plots per row.b
g.map(plt.hist, 'comb', bins=bin_edges);
```

# Example 3

Creating a facet grid that plots information for each transmission type in decreasing order of combined fuel effeciency.

```python
# Finding the combined fuel efficiency for each transmission type
group_means = df[['trans', 'comb']].groupby(['trans']).mean()

# Selecting only the list of transmission in decreasing order of combined fuel effeciency
group_order = group_means.sort_values(['comb'], ascending=False).index

g = sns.FacetGrid(data=df, col='trans', col_wrap=7, col_order=group_order)
g.map(plt.hist, 'comb');
```
