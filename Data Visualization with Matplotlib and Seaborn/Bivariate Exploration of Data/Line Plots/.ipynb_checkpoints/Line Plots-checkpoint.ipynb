**Line Plots** are used to plot two quantitative variables using a line. The y-axis usually represents the mean of the variable being plotted there. It's like a histogram with an average count instead of a frequency count and point positions instead of bar heights. Lines are used to emphasize relative change; the need for a 0 on the y-axis is not needed. It's also used to emphasize trends on the x-axis. Line plots should not be used when there are nominal values on the x-axis.

Time based plots use line plots a lot, which is very useful for economic data like for stocks and currency. In these cases there is usually one measurement per time, the line isn't connecting averages just raw values across time.

Plots that show changes over time are known as time series plots.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

# Example 1

```python
plt.errorbar(data=df, x='displ', y='comb')
#plt.xticks(rotation=15)
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)')
;
```

The dataframe needs to be adjusted. The plot tales in information in order as it appears rather than taking into account the structure of the dataframe.

# Example 2

First, setting a number of bins where the data will be grouped. Then, setting the bin)edges and computing the square for each bin

```python
bin_edges = np.arange(0.6, 7+0.2, 0.2)
bin_centers = bin_edges[:-1]+0.1
```

Cutting the bin alues into discrete intervals

```python
displ_binned = pd.cut(df['displ'], bin_edges, include_lowest=True)
displ_binned
```

Computing the mean and standard error of the mean for the points in each bin.

```python
comb_mean = df['comb'].groupby(displ_binned, observed=False).mean()
comb_std = df['comb'].groupby(displ_binned, observed=False).std()# observed=False is to remove an error warmomg

plt.errorbar(x=bin_centers, y=comb_mean, yerr=comb_std)
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

The gap in the plot is due to the dataset not containing information about that blank area.

# Example 3

Instead of computing summary statistics on fixed bins, computations can be made on a rolling window using padnas `rolling()` function. Since rolling will make computations on rows sequentially, the values should be sorted.

## Computing statistics in a rolling window

```python
df_window = df.sort_values('displ').rolling(15)

x_winmean = df_window.mean(numeric_only=True)['displ']

y_median = df_window.median(numeric_only=True)['comb']
y_q1 = df_window['comb'].quantile(0.25)
y_q3 = df_window['comb'].quantile(0.75)
```

## Plotting the summary data

Instead of plotting just one line, mutliple lines can be drawn on the plot. The three central quartiles can be plotted together each having their own color.

```python
base_color = 'tab:blue'
line_color = 'darkblue'

plt.scatter(data=df, x='displ', y='comb')
plt.errorbar(x=x_winmean, y=y_median, c="yellow")
plt.errorbar(x=x_winmean, y=y_q1, c='red',linestyle='--')
plt.errorbar(x=x_winmean, y=y_q3,c='black',linestyle='--')

plt.xlabel('Engine Size')
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)')
;
```

# Example 4

Another bivariate application of line plots is to plot the distribution of a numeric variable for different levels of a categorical data. Instead of using violen plots, box plots or facted histograns, multiple lines representing histograms can be ploted over each other in the same plot.

This can be done by using multiple `errorbar()` calls and performing multiple `hist` calls as well as using the `histtype='step'` parameter.

```python
bin_edges = np.arange(-3, df['comb'].max()+1/3, 1/3)


g= sns.FacetGrid(data=df, hue='VClass',height=5)# size is deprecated, use height instead

g.map(plt.hist, 'comb', bins=bin_edges, histtype='step')
g.add_legend()

;
```

Due to overlapping, it can be hard to see some of the category lines on the plot.
