Some univariate plots, like bar charts, rely on the frequency to be in the y-axis. If the y-axis was instead replaced with some other numeric variable (the mean could be used), then it would become a bivariate plot. The same logic can be applied with a histogram.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

# Example 1

```python
sns.barplot(data=df, x='VClass', y='comb')
plt.xticks(rotation=15)
plt.ylabel('Avg. COmbined Fuel Efficiency (mpg)');
```

The `barplot()` function assigns a default estimate of 95% confidence for each bar. This can be changed by using the `errwidth` parameter to determine the confidence level.

## Turning off the whiskers

```python
sns.barplot(data=df, x='VClass', y='comb', errwidth=0) #errwidth should be replaced with err_kws={'linewidth': 0}
plt.xticks(rotation=15)
plt.ylabel('Avg. COmbined Fuel Efficiency (mpg)');
```

## Showing the uncertainty of the variance

```python
sns.barplot(data=df, x='VClass', y='comb', color='tab:orange', errorbar=('ci', 99))
plt.xticks(rotation=15)
plt.ylabel('Avg. COmbined Fuel Efficiency (mpg)');
```

## Changing the whiskers to display the standard deviation

```python
sns.barplot(data=df, x='VClass', y='comb', color='tab:orange', ci='sd') #ci='sd' should be replaced with errorbar='sd'
plt.xticks(rotation=15)
plt.ylabel('Avg. COmbined Fuel Efficiency (mpg)');
```

# Example 2 

The `pointplot()` function can be used to plot averages as points rather than bars. This is useful when the bars don't need to be in reference of 0.

```python
sns.pointplot(data=df, x='VClass', y='comb',errorbar='sd')
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

The lines can be removed by specifying `linestyles=""`

```python
sns.pointplot(data=df, x='VClass', y='comb',errorbar='sd',linestyles="")
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

The focus around the standard deviation can be removed

```python
sns.pointplot(data=df, x='VClass', y='comb')
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

The adapted bar chart is great when the data is binary as the box and violen plots will not be informative

# Example 3

Bringing a few charts together

```python
fig, ax = plt.subplots(1,3, figsize=(20,5))

sns.violinplot(data=df, x='VClass', y='comb', inner=None, ax=ax[0])
ax[0].tick_params(axis="x", rotation=45)

sns.boxplot(data=df, x='VClass', y='comb', ax=ax[1])
ax[1].tick_params(axis="x", rotation=45)

sns.barplot(data=df, x='VClass', y='comb', ax=ax[2])
plt.xticks(rotation=45);
```
