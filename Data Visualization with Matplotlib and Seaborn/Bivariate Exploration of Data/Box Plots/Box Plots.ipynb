**Box Plots** are used to show the relationship between numerical and categorical data, similar to violen plots. They provide summary statistics by showing:

* The median as a line in the center
* First and Third quartile as edges of the upper and lower edges of the box
* Whiskers showing the largest and smallest values in the distribution (not necessarily overall), by default it's 1.5 times the interquartile range in each direction from the edges of the box
* Outliers that are placed outside of the whiskers

# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)
df['VClass'] = df['VClass'].astype(vclasses);
```

# Example 1: Box Plot

```python
sns.boxplot(data=df, x='VClass', y='comb')
plt.xticks(rotation=15)
;
```

# Example 2: Horizontal Box Plot

As with violen plots, `sns.boxplot` infers the direction of the box plot depending on what values it receives for x and y

```python
sns.boxplot(data=df, y='VClass', x='comb');
```
