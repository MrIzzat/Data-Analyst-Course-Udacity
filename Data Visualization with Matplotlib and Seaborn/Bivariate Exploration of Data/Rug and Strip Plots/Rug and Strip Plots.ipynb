**Rug Plots** plot each point as a tick on a single axis. They are used to for quantative data. They are great when added in a join grid with a scatter plot.

**Strip plots** are similar to rug plots in that instead of a tick it uses a dot. It's also similar to swarm plots in that it uses dots but differs in that it's one dimensional (does not use jittering).

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

## Converting the `VClasss` column into an ordered categorical type

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)
df['VClass'] = df['VClass'].astype(vclasses);
```

```python
df.head()
```

# Rug Plots

```python
sns.rugplot(data=df, x='co2')
```

```python
sns.rugplot(data=df, x='comb')
```

```python
g = sns.JointGrid(data=df, x='comb', y='co2')
g.plot_joint(plt.scatter)
g.plot_marginals(sns.rugplot, height=0.3)#Specify how tall each tick should be
;
```

# Strip Plots

Strip plots are similar to violen plots with the `inner = "stick"` or the `inner = "point"` parameters.

```python
plt.figure(figsize = [20, 5])

# left plot: strip plot
plt.subplot(1, 2, 1)
ax1 = sns.stripplot(data=df, x='co2', y='VClass', color='tab:blue')

# right plot: violin plot with inner strip plot as lines
plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='co2', y='VClass', color='tab:blue', inner='stick')
;
```
