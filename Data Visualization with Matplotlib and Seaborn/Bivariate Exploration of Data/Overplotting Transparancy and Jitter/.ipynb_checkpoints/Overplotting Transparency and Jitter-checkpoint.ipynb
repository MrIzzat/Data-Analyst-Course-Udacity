Sometimes scatter plots are hard to read due to the high number points or having values that are discrete in nature. 

Having too many points in a small area is known as **overplotting**. Overplotting can be resolved with:

* **Jitter:** moves the position of each point slightly from it's true value. Is great for handling values that are discrete in nature.
* **Transparency:** makes the points more see through, helps to see where dots are grouped together. Normally, points are see through, but when they overlap they produce a more colorful area, indicating a trend for the distribution or relationship.
* **Sampling:** plotting some of the observations instead of all of them.

With seaborn, jitter and transparency don't affect calculations for the linear regression function.

# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

# Example 1: Jitter

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,5))

sns.regplot(data=df, x='year', y='comb', truncate=False, ax=ax1)#Normal Scatter plot

sns.regplot(data=df, x='year', y='comb', truncate=False, x_jitter=0.3, ax=ax2);
```

# Example 2: Jitter and Transparaency

```python
sns.regplot(data=df, x='year', y='comb', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20})#With transparaency and jitter
;
```

```python
plt.scatter(data=df, x='year', y='comb', alpha=1/20);#With transparency
```
