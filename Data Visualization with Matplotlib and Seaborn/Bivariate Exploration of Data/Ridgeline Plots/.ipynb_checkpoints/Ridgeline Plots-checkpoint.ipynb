**Ridgeline plots** are a series of vertically faceted line plots or density curves. It's often thought as a contrast to the line plot variation where multiple lines were plotten on the same axes with different hues.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format ='retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

## Converting the `VClasss` column into an ordered categorical type

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)
df['VClass'] = df['VClass'].astype(vclasses);
```

# Ridgeline plot

The following a facet grid that will be converted to a ridgline plot

```python
group_means['comb'].max()
```

```python
group_means = df.groupby(['make']).mean(numeric_only=True)
group_order = group_means.sort_values(['comb'], ascending=False).index

g = sns.FacetGrid(data=df, col='make', col_wrap=5, height=2,
                 col_order=group_order)
g.map(plt.hist, 'comb', bins = np.arange(5, group_means['comb'].max()+1, 1))
g.set_titles('{col_name}')
;
```

The ridgline version can be seen below

```python
group_means = df.groupby(['make']).mean(numeric_only=True)
group_order = group_means.sort_values(['comb'], ascending=False).index

g = sns.FacetGrid(data=df, row='make', height=1, aspect=7,
                 row_order=group_order)
g.set(ylim=(0, 0.5))
g.map(sns.kdeplot, 'comb', fill = True)
g.set_titles('{row_name}')
```

Creating overlap between individual subplots. does not look very clean.

```python
group_means = df.groupby(['make']).mean(numeric_only=True)
group_order = group_means.sort_values(['comb'], ascending=False).index

# adjust the spacing of subplots with gridspec_kws
g = sns.FacetGrid(data=df, row='make', height=1, aspect=12,
                 row_order=group_order, gridspec_kws = {'hspace' : -0.001})
g.map(sns.kdeplot, 'comb', fill = True)

g.set(ylim=(0, 0.75))

# remove the y-axes
g.set(yticks=[])
g.despine(left=True)
g.set_ylabels('')

g.set_titles('{row_name}')
;
```

Adding some axeses for each plot

```python
group_means = df.groupby(['make']).mean(numeric_only=True)
group_order = group_means.sort_values(['comb'], ascending = False).index

g = sns.FacetGrid(data=df, row='make', height=1, aspect=12,
                 row_order=group_order, gridspec_kws={'hspace' : -0.2})
g.map(sns.kdeplot, 'comb', fill=True)

g.set(yticks=[])
g.despine(left=True)

g.set(ylim=(0, 0.75))

# set the transparency of each subplot to full
g.map(lambda **kwargs: plt.gca().patch.set_alpha(0))

# remove subplot titles and write in new labels
def label_text(x, **kwargs):
    plt.text(4, 0.02, x.iloc[0], ha = 'center', va = 'bottom')
g.map(label_text, 'make')
g.set_xlabels('comb')
g.set_ylabels('')
g.set_titles('')
;
```
