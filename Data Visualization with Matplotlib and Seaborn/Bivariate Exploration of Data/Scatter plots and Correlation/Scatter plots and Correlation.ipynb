Scatter plots are great for looking at relationships between two quantitative variables, where the values of one variable are on the x axis and the values of the second variable are in the y axis. Each point represents one observation (row). The pattern in the plot can show what kind of relationship there is between the two variables (positive or negative) and the strength.

The strength of linear correlation between two numeric variables can be quantified using a **correlation coefficient**. The Pearson correlation coefficient (r) is the most common coefficient used, where it takes a value between -1 and 1. If the coefficient is positive, then that means the relationship between the two variables is positive, if one increases so does the other. If the coefficient is negative, the the relationship between the two variables is negative, if one decreases so does the other. The higher the magnitude of the coefficient the stronger the relationship between the variables is. A coefficient of 0 indicates there is no or a weak relationship between the two variables.

As seen with ascombe's quartet, plots can have the exact same pearson correlation coefficient but look very different.

Log scaling can be very useful for linear regression because it can transform non-linear trends into linear trends for regression. This kind of linear regression tries to predict  the log(y) from x instead of y.

log(y) = mx + b     instead of     y = mx + b 

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

```python
df.head(10)
```

```python
df.info()
```

# Example 1: Scatter plot showing negative correlation between two variables

## In matplotlib

```python
plt.scatter(data=df, x='displ', y='comb')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

There seems to be an inverse correlation, the lower the combined fuel efficency the more displacement.

## In Seaborn

```python
sns.scatterplot(data=df, x='displ', y='comb')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

Using `regplot()`, a scatter plot can be created with a regresseion function fitting

```python
sns.regplot(data=df, x='displ', y='comb')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

The regression line shows a inverse correlation (negative correlation) between the two variables.

# Example 2: Scatter plot showing a positive correlation between two variables

Sometimes, the line shows shades around the regression line tos how the confidence region for the regression estimation.

```python
sns.regplot(data=df, y='co2', x='displ')
plt.xlabel('Displacement (1)')
plt.ylabel('CO2 emmisions');
```

# Example 3: Plot the regression line on the transformed data

```python
def log_trans(x, inverse=False):
    if inverse:
        return np.power(10,x)
    else:
        return np.log10(x)

sns.regplot(x=df['displ'], y=df['comb'].apply(log_trans))
tick_locs = [10,20,50,100]
plt.yticks(log_trans(tick_locs),tick_locs);
```
