**Swarm plots** are plots that are similar to scatter plots but they plot each data point within a position of a categorical data. It's similar to the violin plot but uses the actual points instead of distribution.

Swarm plots should not be used when there are too many points as there will be distortion and lots of overlapping.

When there are too many points, a good alternative is to use the **strip plot**.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

## Converting the `VClasss` column into an ordered categorical type

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.CategoricalDtype(ordered=True, categories=sedan_classes)
df['VClass'] = df['VClass'].astype(vclasses);
```

```python
plt.figure(figsize = [15, 5])#Very Slow execution

# left plot: violin plot
plt.subplot(1, 3, 1)
ax1 = sns.violinplot(data=df, x='VClass', y='comb', color='tab:blue')
ax1.tick_params(axis="x", rotation=45) 
# center plot: box plot
plt.subplot(1, 3, 2)
ax2 = sns.boxplot(data=df, x='VClass', y='comb', color='tab:blue')
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
ax2.tick_params(axis="x", rotation=45) 

# right plot: swarm plot
plt.subplot(1, 3, 3)
ax3 = sns.swarmplot(data=df, x='VClass', y='comb', color='tab:blue')
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
ax3.tick_params(axis="x", rotation=45) 
;
```
