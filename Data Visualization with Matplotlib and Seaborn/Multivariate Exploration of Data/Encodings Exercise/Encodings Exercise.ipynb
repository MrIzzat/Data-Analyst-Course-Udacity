# Exercise 1: Encoding 3 numeric variables

Let's look at the relationship between fuel mileage in highway settings (`highway`), engine displacement (`displ`), and co2 emissions (`co2`).  
**Use a _scatter plot_ to depict the data, with a color encoding.**

As you work through this, identify:
1. What is the independent variable?
2. What is most appropriate variable for color encoding?
3. Describe the relationships.

```python
plt.scatter(data=df, x='highway', y='displ', c='co2', cmap='viridis_r')
plt.xlabel('Highway Fuel Milage (mpg)')
plt.ylabel('Engine Displacement(1)')
plt.colorbar(label='CO2 (g/mi)')
;
```

1. The engine displacement is the independent variable bcause it determines the milage and the co2 emmisions
2. The `co2` variable is the best variable to color encode because it has a wide range of continuous values that can be approiately expressed through colors.
3. There is a negative relationship between the Engine size and the highway fuel milage. There is a negative relationship between the highway fuel milage adn the co2 emisions. There is a positive relationship between the engine size and co2 emmissions.

# Exercise 2: Color Encoding

Use a box plot to compare the green house gas rating (`ghgScore`) for Premium and Regular gas cars (`fuelType`) for Ford and Chevrolet (`make`).

For this excercise, you will need to cut or mask your data to select only the `fuelType` and `make` of interest.

## Getting the rows that include the two make types

```python
df_plot1 = df[(df['make'] == 'Ford') | (df['make'] == 'Chevrolet')]
```

## Getting the rows with the two fuel types

```python
df_plot2 = df[(df['fuelType'] == 'Premium Gasoline') | (df['fuelType'] == 'Regular Gasoline')]
```

```python
import matplotlib.lines as mlines

plt.figure(figsize=(12,5))


sns.boxplot(data=df_plot1, x='make',y='ghgScore')
sns.boxplot(data=df_plot2, x='fuelType',y='ghgScore')


blue_square = mlines.Line2D([], [], color='tab:blue', marker='s', linestyle='None',
                          markersize=10, label='Make')

orange_square = mlines.Line2D([], [], color='tab:orange', marker='s', linestyle='None',
                          markersize=10, label='Fuel Type')


handles = [blue_square, orange_square]

plt.legend(handles=handles,title='Make',loc='upper right')

;
```

```python
from matplotlib.patches import Patch

plt.figure(figsize=(12,5))


sns.boxplot(data=df_plot1, x='make',y='ghgScore')
sns.boxplot(data=df_plot2, x='fuelType',y='ghgScore')


handles = [
    Patch(facecolor='tab:blue', label='Make'),
    Patch(facecolor='tab:orange', label='Fuel Type')
]


plt.legend(handles=handles,title='Make',loc='upper right')

;

```

```python
df_plot3 = df_plot1[(df_plot1['fuelType'] == 'Premium Gasoline') | (df_plot1['fuelType'] == 'Regular Gasoline')]
```

```python
sns.boxplot(data=df_plot3,x='fuelType', y='ghgScore', hue='make')
;
```

```python
sns.boxplot(data=df_plot3,hue='fuelType', y='ghgScore', x='make')

plt.legend(title='Fuel Type',loc='upper right')
;
```
