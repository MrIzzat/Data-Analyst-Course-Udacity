Faceting is usually done with one row. Utilizing rows allows for adding an extra variable for visualization.

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Dataset

```python
df = pd.read_csv('fuel-econ.csv')
```

## Creating a transmission type column

```python
df['trans_type'] = df['trans'].apply(lambda x:x.split()[0])
```

## Getting a random susbet from the dataset

```python
np.random.seed(2018)

sample = np.random.choice(df.shape[0], 200, replace=False)
df_subset = df.loc[sample]
```

# Plot 1: Scatterplots of Fuel Efficency by Vehicle Class

```python
g = sns.FacetGrid(data=df, col='VClass')
g.map(plt.scatter, 'displ', 'comb')
;
```

# Plot 2: Scatterplot Fuel Efficiency by Vehicle Class and Transmission Type

```python
g = sns.FacetGrid(data=df, col='VClass', row='trans_type', margin_titles=True)
g.map(plt.scatter, 'displ', 'comb')
;
```

# Faceting For Multivariate Data

```python
g = sns.FacetGrid(data=df, col='trans_type', height=7)
g.map(sns.boxplot, 'fuelType', 'highway')



for ax in g.axes.flat:
    ax.set_xticks(range(5))
    _ = ax.set_xticklabels(ax.get_xticklabels(), rotation=15)


;
```

```python
df_fueltypes = df[(df['fuelType'] == 'Regular Gasoline') | (df['fuelType'] == 'Diesel')]

g = sns.FacetGrid(data=df_fueltypes, col='fuelType', row='trans_type', height =7, margin_titles=True)
g.map(sns.scatterplot, 'co2','highway')
;
```
