# What are Pie charts

Pie charts are used to plot relative frequencies for qualitative data. Each category is indicated by a slice of the circle where the area is the relative frequency of that category.

A variant of the pie chart is the donut chart that is similar except that the center of the circle has been removed.

Most of the time, it's better to use a bar chart instead of either the donus or  the pie charts as it's easier to get an estimate of the relative frequencies of each category with it, especially when there are many categories as pie charts don't work too well with that.

Pie charts should be used when:

* It's important to show how the data is broken down into parts
* Only a small number of slices should be plotted

Plotting should start at the 12 o clock position and data marked clockwise based on the realtive frequency, though there are other ways to plot with a pie chart.

# Import Libraries

```python
import pandas as pd

import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('./diamonds.csv')
```

# Plotting a simple Pie chart

```python
sorted_counts = df['clarity'].value_counts()
sorted_counts
```

```python
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90,counterclock=False)
plt.axis('square'); #Ensures that the scaling of the plot is equal on both the x and y axes to prevent the chart having an oval-shape
```

## Custom labels

```python
labels = ['Slight occulsion 1', 'Very slight occulsion 2', 'Slight occulsion 2', 'Very slight occulsion 1'
          , 'Very Very slight occulsion 2', 'Very Very slight occulsion 1', 'No internal flaws','Incl']

plt.pie(sorted_counts, labels=labels, startangle=90,counterclock=False)
plt.axis('square');
```

# Donut Plot

A donut plot can be created by setting the parameter value of the `wedgeprops` to have a width less than 1

```python
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90,counterclock=False, wedgeprops = {'width' : 0.3})
plt.axis('square');
```
