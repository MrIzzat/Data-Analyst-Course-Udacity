Visualizations created can give insights into data statistics that can't be gotten from descriptive statistics. Such as if the model is skewed or if it's modal type. They can also show interesting locations for further investigation.

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%config InlineBackend.figure_format='retina'
```

# Load Dataset 

```python
df = pd.read_csv('diamonds.csv')
```

It's good to keep into account the number of modes, skew and presence of outliers.

# Histogram of `price` Data

```python
sns.histplot(data=df, x='price');
```

The `price` data is skewed to the right.

# Axis Limits

The axis can be limited to only show a certain number of values. `plt.xlim()` can be used to limit the x axis and `plt.ylim()` to limit the y-axis.

```python
sns.histplot(data=df, x='price')
plt.xlim(0,2500);
```

```python
bins = np.arange(0 ,df['price'].max()+50,50)
plt.hist(data=df, x='price', bins=bins)
plt.xlim(1000,5000);
```
