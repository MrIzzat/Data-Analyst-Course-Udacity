# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv('die-rolls.csv')
```

```python
df.head(10)
```

# Shifting edges of the bars

Shifting the edges of the bars can remove ambiguity in the case of discrete data

```python
plt.figure(figsize = [20,5])

plt.subplot(1,2,1)
bin_edges = np.arange(2,12+1.1,1)
plt.hist(data=df,x='Sum',bins=bin_edges)
plt.xticks(np.arange(2,12+1,1))

plt.subplot(1,2,2)
bin_edges = np.arange(1.5,12.5+1,1)
plt.hist(data=df,x='Sum',bins=bin_edges)
plt.xticks(np.arange(2,12+1,1));
```

# Adding gaps between individual bars

The `rwidth` paramter can be used to change the proportion of the bin widths filled by each bar. Gaps between bars make the data look more discrete. Plotting data like makes it seem like the data is ordinal-type instead of quantitative.

```python
bin_edges = np.arange(1.5,12.5,+1)
plt.hist(data=df, x='Sum', bins = bin_edges, rwidth = 0.7)
plt.xticks( np.arange(2, 12+1, 1));
```
