# What is absolute vs relative frequency?

The absolute frequency refers to the number of data points in each category.

The relative frequency refers to the proportion of data points in each category.



![](./Example.png)








# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

# Load dataset

```python
df = pd.read_csv('./diamonds.csv')
```

```python
df.head()
```

# Absolute Frequency

The absolute frequency can be plotted right away with Seaborn with the `countplot()` function as in the `Bar Chart` notebook. Or the counts can be calculated using the `value_counts()` function.

```python
order = df['clarity'].value_counts().index
```

```python
sns.countplot(data=df, y='clarity', color='tab:blue', order=order);
```

# Relative Frequency

Since the proportions are going to be used, the `countplot()` function won't work. Instead, the `barplot()` function should be used. The proportional data can be calculated by using the `value_counts()` function in pandas with the `normalize=True` parameter.

## Calculating proportions

```python
proportions = df['clarity'].value_counts(normalize=True)
```

```python
proportions
```

## Renaming columns 

To make the axis names cleaner

```python
proportions = proportions.reset_index(drop=False)#Converts a series into a data frame, which creates a new column called 'index'
```

```python
proportions = proportions.rename(columns={'index':'clarity'})
```

```python
proportions
```

## Plotting the proportions

```python
sns.barplot(data=proportions,y='clarity',x='proportion');
```

## Manually getting the normalized data

```python
total_count = df['clarity'].value_counts().sum()
clarity_counts = df['clarity'].value_counts()
clarity_counts = clarity_counts/total_count
clarity_counts
```

# Absolute and Relative Frequency in the Same Plot

Instead of plotting the relative frequency onto it's own plot, text annotations can be used to add the relative frequency of each absolute frequency bar.

```python
plt.figure(figsize=(13,5))
sns.countplot(data=df, y='clarity', color='tab:blue', order=order)

total_count = df['clarity'].value_counts().sum()

for i,count in enumerate(df['clarity'].value_counts()):

    text_annotation = f'{100*count/total_count:.1f}%'

    plt.text(count+1, i, text_annotation, va='center')
```
