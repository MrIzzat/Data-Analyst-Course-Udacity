# What are histograms?

A **Histogram** is excellent for plotting quantitative variables. It's similar to the bar chart in that it has feature values on the x axis and numeric values on the y axis. The difference between them is that bar chart bars are used to plot single values while bars in histograms are used to plot ranges.
The range of each bar is referred to as **bin size**. Bin sizes that are too large may lose information about distribution of different data. Bin sizes that are too small may add noise that can distract from trends in the data. It's good to experiment with bins sizes to achieve the desired results.

# Import Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Load dataset

```python
df = pd.read_csv('./diamonds.csv')
```

# Plots

## Default Histogram

```python
plt.hist(data=df, x='x')
```

The data that was outputted with the graph is the histogram bin size per each bin, this value can be specfied with an argument.

The data is not a uni-form distribution, a smaller bin size may prove more useful.

By default, the `hist()` function divides the data into 10 bins.

## Histogram with specific number of bins

The number of bins can be specified using the `bins=` argument

```python
plt.hist(data=df, x='x', bins=25);
```

## Histogram with dynamic number of bins

The specific number of bins and sizes can be specified used the `np.arange()` function. The first argument is the leftmost bin edge, the second is the upper limit and the third argument is the bin width.

```python
bins = np.arange(0, df['x'].max()+0.25, 0.25)#the extra +0.25 in the second argument is because arange() 
                                             #will only return values strictly less than the upper limit
plt.hist(data=df, x='x', bins=bins);
```

`np.around()` can be used to programmatically find the leftmost bin's value

```python
bins = np.arange(np.around(df['x'].min()), df['x'].max()+0.25, 0.25)
plt.hist(data=df, x='x', bins=bins);
```

## Histograms with Seaborn

The syntax of the histogram function in seaborn is: `seaborn.histplot(data, x,y, bins, kde, stat, hist_kws)`

```python
sns.histplot(data=df, x='x',bins=bins);# bins was created previously
```

Seaborn can renormalize data to percentages using the `stat='percent'` argument. It can also find the density distribution in the histogram with the `kde=True` argument.

```python
sns.histplot(data=df, x='x',bins=bins, stat='percent', kde=True);
```

## Plotting two histograms side-by-side

Viewing multiple histograms at once can be achieved by using the `plt.subplot()` and `plt.figure()` functions

```python
plt.figure(figsize = [20,5])

plt.subplot(1,2,1)# 1 row, 2 columns, subplot 1

# left histogram
bins = np.arange(0,df['x'].max()+1,1)
plt.hist(data=df, x='x',bins=bins);

# right histogram

plt.subplot(1,2,2) #1 row, 2 columns, subplot 2
bins =  np.arange(0,df['x'].max()+0.05,0.05)
plt.hist(data=df, x='x', bins=bins);
```

The first two arguments refer to how many rows and columns there should be in a plot. The third argument refers to which subplot number is being referenced. Multiplying the number of rows by the number of columns indicates how many total subplots there are, this can be beneficial when estimating the number of rows and columns to use when plotting multiple sub plots.

## Plotting multiple variables using the hue argument in seaborn

Unpivoting the data frame using the `df.melt()` function converts data in to a long format and allows for the plotting of multiple variables.

```python
df = df[df['y'] < 11. ] #Removes outliers

df['index'] = df.index

df_long = df.melt(id_vars=['index'], value_vars = ['x', 'y'], var_name='dimension', value_name= 'size')

df_long.shape, df_long.head()
```

```python
sns.histplot(data=df_long, x='size', hue='dimension',bins=61, element='step'); #element='step' removes internal bin lines'
```

```python
sns.histplot(data=df_long, x='size', hue='dimension',bins=61);
```
