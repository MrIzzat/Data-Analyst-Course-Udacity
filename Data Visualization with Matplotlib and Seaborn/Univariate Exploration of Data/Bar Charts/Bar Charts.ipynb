# What are bar charts

Bar charts are very useful for visualization of the distribuation of a qualitative data type.

A bar chart is usually uses the x-axis for each level of the categorical labels (at a unique x position) and the y-axis for the frequency of each category. An example of using a bar chart is counting the number of males to females in a particular field. To avoid any distortions, the base line should start from zero. If the plot does not start from zero, it could make one bar seem proportiantely much larger than the other bar.

For nominal data, it's recommended to sort the bars based on frequency, with the most common being on the left, which makes it see which category is most common and how the other categories compare.

For ordinal data, the bars should not be sorted based on frequency. Instead, the inherent order of the data should be maintained, as it's the more important information to convey. It can also give a better idea of the distribution.

# Import Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

# Load Dataset

```python
df = pd.read_csv('./diamonds.csv')
```

# Viewing Basic Information 

```python
df.shape
```

```python
df.head()
```

# Count Plot in Seaborn

Counting the number of diamonds for each cut

```python
sns.countplot(data=df, x='cut');
```

## Modifying the color

Seeing Seaborn's color palette

```python
sns.color_palette()
```

## Choosing and applying the color

```python
color = sns.color_palette()[4]
```

```python
sns.countplot(data=df, x='cut',color=color);
```

## Selecting a color by name

The names of the colors can only be chosen from the color palette. Anything outside the palette will cause an error.

```python
sns.countplot(data=df, x='cut', color="tab:blue");
```

## Setting a custom RGB color value

```python
sns.countplot(data=df, x='cut',palette=sns.color_palette(['#00119B'],n_colors=5),hue='cut');
# n_colors=5 refers to the 5 bars in the bar chart, it removes a warning
```

# Ordering the plot

The `order=` parameter can be used to order the values in a certain order

```python
order = df['cut'].value_counts().index
```

```python
sns.countplot(data=df,x='cut',color='tab:blue',order=order);
```

# Rotating axis labels and bars

The labels on the x axis can be rotated using matplotlib's `xticks()` function. *I think a similar idea can be applied to the y axis labels*

```python
sns.countplot(data=df, x='cut', color='tab:blue',order=order)
plt.xticks(rotation=15);
```

If the countplot is called again any previous modifications will not appear again

```python
sns.countplot(data=df, y='cut', color='tab:blue',order=order);
```

# Bar Charts using Matplotlib

## Getting the frequency of each cut

```python
count = df['cut'].value_counts(sort=False)
count
```

## Plotting the results

```python
plt.bar(count.index,count)

plt.xlabel('cut')
plt.ylabel('count')

plt.show()
```

# Plotting the diamond colors and sorting based on frquency

As mentioned before, will be using the `order()` function. The order can be calculated manually (statically) or programmatically (dynamically)

## Static ordering

Have to manual order the columns

```python
sns.countplot(data=df, x='color', color='tab:blue', order=['G','E','F','H','D','I','J']);
```

## Dynamic ordering

```python
counts = df['color'].value_counts()
counts
```

```python
sns.countplot(data=df, x='color', color='tab:blue', order=counts.index);
```
