Each visualization is a `Figure` object that contains one or more `Axes` objects, each Axes object contains a number of other elements that are used to represent the plot.

A normal line like `plt.hist(data=df, x='x')` creates these objects implicitly.

# Load Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format="retina"
```

# Load Dataset

```python
df = pd.read_csv('./diamonds.csv')
df.shape
```

```python
df.head()
```

# Using `Figure.add_axes()` and `axes.hist()`

`figure()` is used to create a `Figure` object.

```python
fig = plt.figure()
```

`add_axes()` canb used to create new axes in the newly created figure object. It uses a list as a parameter to represent the dimensions: `[left, bottom, width, height]`. The quantities are in fractions of the figure.

```python
fig = plt.figure()

ax = fig.add_axes([.125,.125,.775,.755])
ax.hist(data=df,x='x')
```

# Using axes with `seaborn.countplot()`

```python
fig = plt.figure()
ax = fig.add_axes([.125,.125,.775,.755])
sns.countplot(data=df,x='cut', color='tab:blue',ax=ax)
```

# Sub-plot

Multiple figures can be plotted next to each other with the use of the `subplot()` function. It takes the arguements `(number of rows, number of columns, subplot number)`, where the total number of possible subplots can be found by multiplying the number of rows by the number of columns.

```python
plt.figure(figsize = [20,5])

plt.subplot(1,2,1) # 1 row, two columns, subplot 1
bins = np.arange(0,df['x'].max()+0.25, 0.25)
plt.hist(data=df, x='x', bins=bins)

plt.subplot(1,2,2)
bins = np.arange(0,df['x'].max()+0.05, 0.05)
plt.hist(data=df,x='x',bins=bins);
```

## Without Specifying subplots

```python
plt.figure(figsize = [20,5])


bins = np.arange(0,df['x'].max()+0.25, 0.25)
plt.hist(data=df, x='x', bins=bins)


bins = np.arange(0,df['x'].max()+0.05, 0.05)
plt.hist(data=df,x='x',bins=bins);
```

# Additional Techniques

`plt.gca()` can be used to retrieve the current axes so that an axes object does not need to be assigned to a variable as soon as it's created. A list of all of the axes in a figure can be retreived by used `fig.get_axes()`. `fig.add_subplots()` can be used in addition to `plt.subplots()` to create subplots.

## Using subplots with seaborn

```python
fig, ax = plt.subplots(1,2, figsize=(10,3))
sns.countplot(data=df, x='cut',color='tab:blue', ax=ax[0])
sns.countplot(data=df, x='cut',color='tab:red', ax=ax[1]);
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,3))
sns.countplot(data=df, x='cut',color='tab:blue', ax=ax1)
sns.countplot(data=df, x='cut',color='tab:red', ax=ax2);
```
