Waffle plots are an alternative to univariate categoral plots (like the pie chart). It uses a 10x10 grid where each square represents 1% of the data. A square is colored by the color of the category it represents.

# Import Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format="retina"
```

# Load dataset

```python
df = pd.read_csv('diamonds.csv')
```

# Function to decicde which category gets number of blocks

```python
def perecentage_block(df,var):
    
    percentages = 100 * df[var].value_counts() / df.shape[0] #Gets the percentage of each category
    counts = np.floor(percentages).astype(int)# Makes sure each percentage is stored as an integer
    decimal = (percentages - counts).sort_values(ascending=False)#Gets the left over percentages

    rem = 100 - counts.sum() #Finds the number of left over squaraes

    # print(percentages)
    # print( decimal.index[:rem]) #For visualizations
    # print(counts.sum())
    # print(decimal)
    
    for cat in decimal.index[:rem]: #decimal.index[:rem] means to take the categories with the most left over percentages
        counts[cat] +=1             # Add the left over squares to the categories with the most left over perentages
                                    # The loop only goes until the extra left over squares are allocated to a category

    return counts
```

# Other Variant of Waffle plot

Instead of using relative frequency, absolute frequency can be used where each square represents a certain number instead of a percentage.

```python
# Each box represents five counts
waffle_counts = (df['cut'].value_counts()/1000).astype(int)

prev_count = 0

for cat in range(waffle_counts.shape[0]):

    blocks = np.arange(prev_count, prev_count + waffle_counts.iloc[cat])
    
    x = blocks %10
    y = blocks //10

    plt.bar(x=y, height = 0.8, width = 0.8, bottom=x)#Flipped the x and y values
    prev_count += waffle_counts.iloc[cat]

plt.bar(7.5, 0.8, 0.8, 2, color='white', edgecolor = 'black', lw=2)
plt.text(8.1, 2.4, '= 1000 data points', va= 'center')

plt.legend(waffle_counts.index, bbox_to_anchor = (0.8, 0.5), loc=6)
plt.axis('off')
plt.axis('square');
```

Somtimes, icons can be used instead of squares in waffle plots but they should be used sparingly to avoid chart junk.
