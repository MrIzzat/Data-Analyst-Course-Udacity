The **Kernel Density Estimation (KDE)** is used to estimate the probability density function of a variable. Each observation represents part of the area of the probability density function, combining these areas together yields the final density curve. The normal-distribution kernel is usually the default, which means it assumes each area is part of a gaussian (normal) distribution.

# Example 1: Plot the Kernel Density Estimatation (KDE)

```python
sns.histplot(data=df, x='price', kde=True);
```

# Example 2: Demonstrating kdeplot() and rugplot()

```python
data = [0.0, 3.0, 4.5, 8.0]
fig, (ax1,ax2) =  plt.subplots(1,2,figsize=(12,5))

#kde with rugplot
sns.kdeplot(data, ax=ax1)
sns.rugplot(data, ax=ax1, color='r')

#kde with narrow bandwidth to show invidual proabilitylumps
sns.kdeplot(data,ax=ax2, bw_adjust=0.2)# The bw_adjust parameter adjusts the size of the bandwidth, it's like changing the bin size and numbers in histograms
sns.rugplot(data,ax=ax2, color='r');#Bandwidth that is too small may make data seem more noisy than it is, too big and it will smooth out important information
```
