# Visualizing the missing data

Seaborn's `barplot()` function can be used to for this.

```python
nacolumns = df.drop(['Date','Temperature','Fuel_Price'],axis=1).isna().sum()
```

```python
nacolumns
```

```python
plt.figure(figsize=(13,5))
sns.barplot(y=nacolumns.index.values,x=nacolumns).set_title("Missing data")



for i in range(nacolumns.shape[0]):

     count = nacolumns.iloc[i]

     plt.text(count, i, count)
```
